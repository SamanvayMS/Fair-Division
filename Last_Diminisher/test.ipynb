```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
shapefile_path = '../Resources/tl_2012_us_state/tl_2012_us_state.shp'
```

```python
us_map = gpd.read_file(shapefile_path)
alaska = us_map[us_map['NAME'] == 'Alaska']
hawaii = us_map[us_map['NAME'] == 'Hawaii']
```

```python
us_map.NAME.unique()
```

```python
teritories = ['Puerto Rico', 'Guam', 'United States Virgin Islands','American Samoa','Commonwealth of the Northern Mariana Islands', 'Alaska', 'Hawaii']
```

```python
def drop_teritories(us_map, teritories):
    for teritory in teritories:
        us_map.drop(us_map.index[us_map['NAME'] == teritory], inplace=True)
    return us_map
```

```python
us_map = drop_teritories(us_map, teritories)
```

```python
fig, ax = plt.subplots(3,1,figsize=(15,15))
us_map.plot(ax=ax[0])
alaska.plot(ax=ax[1])
hawaii.plot(ax=ax[2])
for axs in ax:
    for spine in axs.spines.values():
        spine.set_visible(False)
    axs.xaxis.set_ticks([])
    axs.yaxis.set_ticks([])
```
